```python
'''Создание класса с помощью type'''

def method(self):
    return self ** 2

MyCls = type('MyCls', (object,), {'method': method})
m1 = MyCls.method(5)
print(m1)

# эквивалентно обычному определению класса
class MyCls2:
    def method(self):
        return self ** 2

m2 = MyCls2.method(5)
print(m2)
```
