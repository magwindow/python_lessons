```python
def fibonacci_sequence(L):
    '''Последовательность Фибоначчи'''
    if len(L) < 2:
        return L
    fib_sequence = L[:2]
    for i in range(2, len(L)):
        next_num = fib_sequence[i-1] + fib_sequence[i-2]
        fib_sequence.append(next_num)
    return fib_sequence

L = [0, 1, 1, 1, 1, 1]

fib_sequence = fibonacci_sequence(L)
print(fib_sequence)  
```

```python
def transform_number(num: int) -> int:
    '''
    Преобразовывет последовательность чисел в сумму
    первой и последней цифры. 
    Пример: 143562 -> 3108 (1+2, 4+6, 3+5)
    '''
    digits = list(str(num))               # превращаем число в список строковых цифр
    result = []

    half = len(digits) // 2
    for i in range(half):     # перебираем половину списка
        left = int(digits[i])             # цифра с начала
        right = int(digits[-(i + 1)])     # цифра с конца
        result.append(str(left + right))  # суммируем и записываем как строку
        
    # если нечётное количество цифр — добавляем среднюю
    if len(digits) % 2 == 1:
        middle = digits[half]
        result.append(middle)

    return int(''.join(result))           # собираем строку обратно в число

print(transform_number(143562))  # → 3108
print(transform_number(123456))  # → 777
print(transform_number(987654))  # → 131313
print(transform_number(1234))
print(transform_number(12345))   # → 663  (1+5, 2+4, середина 3)
print(transform_number(123))
```

```python
from fractions import Fraction

a = Fraction (6, 4)
b = Fraction (7, 12)
print (a + b)
print (a * b)

# Получение числителя/знаменателя
c = a * b
print(c.numerator)
print(c.denominator)

# Ограничение значения знаменателя
print(c.limit_denominator(8))

# Преобразование из float в дробь
x = 5.75
y = Fraction(*x.as_integer_ratio())
print(y) 
```

```python
'''Работа со списками'''

# Создание вложенных списков
lst = [[i] for i in range(3)]
print(lst)

# Элементы списка в обратном порядке
lst = [1, 2, 3, 4, 5, 6, 7]
print(lst[::-1])

# Изменение элементов списка с помощью среза
lst = [1, 2, 3, 4, 5, 6, 7]
lst[1:3] = [8, 9]
print(lst)

# Конкатенация списков
lst = [1, 2, 3, 4, 5, 6, 7]
lst2 = [10, 11, 12]
lst3 = lst + lst2
print(lst3)

# Если нужно добавить элементы в текущий список,
# можно использовать оператор +=
lst = [1, 2, 3, 4, 5, 6, 7]
lst += [10, 11, 12]
print(lst)
```
